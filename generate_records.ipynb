```python
from faker import Faker
import pandas as pd

# Create Faker instance
fake = Faker()

# Sample employee data
sample_data = {
    'Name': fake.name(),
    'Title': fake.job(),
    'Department': fake.random_element(elements=('HR', 'IT', 'Finance', 'Marketing')),
    'Salary': fake.random_int(min=40000, max=100000),
    'Email': fake.email(),
    'Phone': fake.phone_number(),
    'Address': fake.address(),
}

# Number of records to generate
num_records = 1000

# Generate employee records
employee_records = []
for _ in range(num_records):
    record = {
        'Name': fake.name(),
        'Title': fake.job(),
        'Department': fake.random_element(elements=('HR', 'IT', 'Finance', 'Marketing')),
        'Salary': fake.random_int(min=40000, max=100000),
        'Email': fake.email(),
        'Phone': fake.phone_number(),
        'Address': fake.address(),
    }
    employee_records.append(record)

# Create a DataFrame from the generated records
df = pd.DataFrame(employee_records)

# Save the DataFrame to a CSV file
df.to_csv('employee_data.csv', index=False)
```
